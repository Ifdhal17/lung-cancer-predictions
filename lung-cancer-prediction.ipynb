**Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
level_counts = df['level'].value_counts()
level_mapping = {0: 'Low', 1: 'Medium', 2: 'High'}

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(level_counts, labels=[f'{level_mapping[label]} ({count})' for label, count in level_counts.items()], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Levels')
plt.show()
```

```python
# create a correlation matrix
plt.figure(figsize=(20, 20))
sns.heatmap(df.corr(), annot=True, cmap='rocket', linewidths=2)
plt.show()
```

```python
# create each of the reg plots
fig, axes = plt.subplots(ncols=4, nrows=6, figsize=(20, 20))
axes = axes.flatten()
# generate colors
colors = sns.color_palette('husl', len(df.columns))

for i, col in enumerate(df.columns):
    sns.regplot(x=col, y='level', data=df, lowess=True , color=colors[i], ax=axes[i])
    axes[i].set_title(f'level vs {col}')
plt.tight_layout()
plt.show()
```

we can see the correlation in each of scatter plot in form of linear regression

```python
# Create box plots for each column
fig, axes = plt.subplots(ncols=4, nrows=6, figsize=(20, 20))
axes = axes.flatten()
for i, col in enumerate(df.columns):
    sns.boxplot(x='level', y=col, data=df, ax=axes[i])
    axes[i].set_title(f'level vs {col}')
plt.tight_layout()
plt.show()
```

as we can see alcohol use, genetic, smoking, even as a passive smoker have a strong correlation with the level of lung cancer. 

```python
from scipy.stats import norm

# create a histogram for each column
fig, axes = plt.subplots(ncols=4, nrows=6, figsize=(20, 20))
axes = axes.flatten()
i = 0

for k, v in df.items():
    mean, std = norm.fit(v)
    sns.histplot(v, ax=axes[i], color=colors[i % len(colors)], kde=True)
    axes[i].text(0.95, 0.95, f'mean: {mean:.2f}\nstd: {std:.2f}', transform=axes[i].transAxes, 
                 verticalalignment='top', horizontalalignment='right', 
                 bbox=dict(facecolor='white', alpha=0.5))
    axes[i].set_title(k)
    i += 1
plt.tight_layout()
plt.show()
```
